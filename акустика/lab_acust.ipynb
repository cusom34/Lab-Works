```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def index_of_closest_point(mass, point):
    min_dist=max(mass)
    min_index=None
    for i in range(len(mass)):
        if abs(point-mass[i])<min_dist:
            min_dist=abs(point-mass[i])
            min_index=i

    # print(min_index, mass[min_index])
    return min_index
```

```python
def plot(x, y, title):
    plt.grid()
    plt.title(title)
    plt.xlabel("Гц")
    plt.ylabel("мВ", rotation=0)
    
    ax = plt.gca()
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    ax.xaxis.set_label_coords(1.03, 0)
    ax.yaxis.set_label_coords(0, 1.03)
    #построение на плоскости
    
    plt.plot(x, y,marker = ".")
    
```
