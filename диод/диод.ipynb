Прямая Ричардсона
ln Jₛ − 2 ln T = ln( Sₐ ⋅ A) − e Φ/kT
   или  
   ln Jₛ − 2 ln T = f(1/T)


Я строил по инфе из задания 4. Там как раз известны и токи накала, и анода, что в целом и дает возможность строить прямые Ричардсона


Федор Сарафанов в своей работе утверждает, что 𝑇(𝑖нак ) = 970.52 + 996.208 · 𝑖нак

```python
def temp(i_n):
  return 970.52+996.208*i_n
```

```python
k = 1.38e-23
scale_to_eV = 1e19/1.6
def plot_Richardson(ia, i_n, color="blue", title="ВАХ", x_label="A", y_label="B", label = ""):
  ia=np.array(ia)
  i_n=np.array(i_n)
  y_ax=[np.log(ia[i])-2*np.log(temp(i_n[i])) for i in range(len(ia))]
  x_ax=1/temp(i_n)
  plot(x_ax, y_ax, color=color, title="",  y_label="ln Jₛ − 2 ln T", x_label="1/T", label = label)
  print(color ,"Эффективная работа выхода = ",scale_to_eV*k*(y_ax[len(x_ax)//2]-y_ax[len(x_ax)//2+1])/(x_ax[len(x_ax)//2+1]-x_ax[len(x_ax)//2]),'эВ\n')
  plt.xscale('log')
  plt.grid(which='major')
  plt.minorticks_on()
  plt.grid(which='minor', alpha=0.3)
  plt.legend(loc="best")
```

```python
plot_Richardson(ia1, in1, label = "Ua = 15 В")
plot_Richardson(ia2, in2, color="g",label = "Ua = 60 В")
plot_Richardson(ia3, in3, color="r",label = "Ua = 100 В")
```
